-column 2 ignore any values which are not 3 or 4

-column 3 is in nanoseconds

To find the timestamp, simply sum

---  (column1 * e-7 + column2 * 3.2e-11) = ts

Next will be a demonstration a simple awk script to bin the single photon count data in order to perform FRET efficiency analysis.

````python
!head -n 1000 ../auxiliary/0.1M\ gdmhcl\ 2.csv | awk 'BEGIN{precision=10^1}; {if (NR > 2){x=int(($1 * 10^-6 + $3 * 10^-9)*precision); y[x]++}}; END{for (j in y) print j, y[j]}' | sort -n > output.txt
````

Now a python implementation including logic to handle both channels.

````python
inputFilename = '../auxiliary/0.1M gdmhcl 2.csv'
precision=pow(10,1) #adjust precision of bins to your liking
channelOne = [0] * 1000000 #quickly defining an arbitrarily large array
channelTwo = [0] * 1000000
with open(inputFilename) as byLine:
    next(byLine) #skip headers
    next(byLine)
    for line in byLine:
        spl = line.split()
        for i in range(3):
            spl[i] = float(spl[i]) #convert from scientific notation to float
        x = int((spl[0] * pow(10,-7) + spl[2] * 3.2 * pow(10,-11))*precision) #see given formula above
        if spl[1] > 2.99 and spl[1] < 3.01: #ignore column 2 values which are not 3 or 4 as per instructions
            channelOne[x] += 1
        elif spl[1] > 3.99 and spl[1] < 4.01:
            channelTwo[x] += 1

for i in range(len(channelOne)-1):
    if channelOne[i] > 0 or channelTwo > 0: #ignore extra items in arbitrarily large array
        print(i, channelOne[i], channelTwo[i]) #print our data

    
````
